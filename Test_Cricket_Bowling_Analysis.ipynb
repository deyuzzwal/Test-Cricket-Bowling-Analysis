# Number of players played for ICC

```python
file = df.copy()

display(file.head())
```

```python
# Rearranging the columns name
new_col_sequence = ['Player', 'Country', 'Matches', 'Innings', 'Wickets', 'Runs', 'Average',
                    'Economy', 'Strike_Rate', 'Five_Wickets_Haul', 'Ten_Wickets_Haul', 'Span']
df = df[new_col_sequence]

display(df.head())
```

```python
def icc_playing_check(xl):
    if "ICC" in xl:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_ICC'] = df['Country'].apply(icc_playing_check)

display(df.head(5))
```

```python
df['played_for_ICC'].value_counts()
```

- The upwards data statistics showed that 5 players played for ICC and 74 players did not play for ICC.

 # Number of Australian Bowlers are present in this dataset

```python
def Australian_bowler_check(xl):
    if "AUS" in xl:
        return "Yes"
    else:
        return "No"
```

```python
df['played_for_Australia'] = df['Country'].apply(Australian_bowler_check)

display(df.head(5))
```

```python
df['played_for_Australia'].value_counts()
```

- The earlier results showed that 18 Australian players participated in test cricket, according to the dataset.

 # Number of Bangladeshi Bowlers are present in this dataset

```python
df['played_for_Bangladesh'] = df['Country'].apply(Bangladeshi_bowler_check)

display(df.head(5))
```

```python
# Span column splitting depends on the "-"
df_span = df['Span'].str.split("-", expand=True)

# conccatination the new data frame with the old data frame
df = pd.concat([df, df_span], axis=1)

# renaming the newly created column names
df = df.rename(columns={0: "start_year",
                        1: "end_year"})

# removing the span column
df = df.drop("Span", axis=1)

display(df.head())
```

```python
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

print(df.info())
```
